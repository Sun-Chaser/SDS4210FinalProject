```python
# SIC Division Bar Chart Counts
import numpy as np
import pandas as pd

def sic_to_division(series):
    s = pd.to_numeric(series, errors="coerce")  # convert to int, NA → NaN

    conditions = [
        s.between(100, 999),
        s.between(1000, 1499),
        s.between(1500, 1799),
        s.between(2000, 3999),
        s.between(4000, 4999),
        s.between(5000, 5199),
        s.between(5200, 5999),
        s.between(6000, 6799),
        s.between(7000, 8999),
        s.between(9100, 9729),
        s.between(9900, 9999),
    ]

    choices = [
        "Agriculture/Forestry/Fishing",
        "Mining",
        "Construction",
        "Manufacturing",
        "Transportation/Public Utilities",
        "Wholesale Trade",
        "Retail Trade",
        "Finance/Insurance/Real Estate",
        "Services",
        "Public Administration",
        "Nonclassifiable",
    ]

    return np.select(conditions, choices, default="Other")

cleaned_df["industry_division"] = sic_to_division(cleaned_df["sic"])
counts = cleaned_df["industry_division"].value_counts().sort_values(ascending=False)

import matplotlib.pyplot as plt

counts.plot(kind="bar", figsize=(10,5))
plt.title("Number of Companies per SIC Industry Division")
plt.ylabel("Count")
plt.xlabel("Industry Division")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Map for country names
country_name_map = {
    "US": "United States",
    "CA": "Canada",
    "GB": "United Kingdom",
    "IE": "Ireland",
    "SG": "Singapore",
    "BM": "Bermuda",
    "AU": "Australia"
}

# Extract country code from loc
country_series = cleaned_df["loc"].str.split("-").str[0]

# Convert to full names
country_names = country_series.map(country_name_map).fillna("Unknown")

# Count companies
country_counts = country_names.value_counts().to_frame(name="company_count")
country_counts.index.name = "country"
country_counts = country_counts.reset_index()

plt.figure(figsize=(10, 6))
plt.bar(country_counts["country"], country_counts["company_count"])
plt.title("Company Density by Country")
plt.xlabel("Country")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Identify quarterly income columns
income_cols = [col for col in cleaned_df.columns if "Q" in col]

# Per-company mean income
mean_income = cleaned_df[income_cols].mean(axis=1, skipna=True)

# Country code → country name
country_series = cleaned_df["loc"].str.split("-").str[0]
country_names = country_series.map(country_name_map).fillna("Unknown")

# Combine for grouping
tmp = pd.DataFrame({
    "country": country_names,
    "mean_income": mean_income
})

# Compute country-level mean
country_income = (
    tmp.groupby("country")["mean_income"]
       .mean()
       .reset_index()
       .sort_values("mean_income", ascending=False)
)

plt.figure(figsize=(10, 6))
plt.bar(country_income["country"], country_income["mean_income"])
plt.title("Mean Quarterly Income by Country")
plt.xlabel("Country")
plt.ylabel("Average Quarterly Income")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
